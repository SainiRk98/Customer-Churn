```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

#from the given pie chart we can conclude that 26.54% of our customers have churned out.

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Normalize to get the percentage of each category
percentage_data = count_data.divide(count_data.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
percentage_data.plot(kind='bar', stacked=True, figsize=(6, 6))

# Add labels, title, and axis labels
plt.title('Churn by SeniorCitizen (Percentage)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper left', labels=['No Churn', 'Churn'])

# Show the plot
plt.show()
```

#compartive a greater percentage of people in senior citizen category have churned

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrame (replace this with your actual DataFrame)
# Assuming your DataFrame is df
data = {
    'PhoneService': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'MultipleLines': ['Yes', 'No', 'No', 'Yes', 'Yes'],
    'InternetService': ['DSL', 'Fiber optic', 'Fiber optic', 'DSL', 'DSL'],
    'OnlineSecurity': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'OnlineBackup': ['No', 'Yes', 'Yes', 'No', 'No'],
    'DeviceProtection': ['Yes', 'Yes', 'No', 'No', 'Yes'],
    'TechSupport': ['No', 'Yes', 'Yes', 'No', 'No'],
    'StreamingTV': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'StreamingMovies': ['Yes', 'Yes', 'No', 'Yes', 'No'],
}

df = pd.DataFrame(data)

# List of columns you want to create countplots for
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the matplotlib figure
fig, axes = plt.subplots(3, 3, figsize=(15, 15))  # 3x3 grid of subplots

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through each column and create a countplot in each subplot
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel(column)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

#In most cases, the "Yes" category has a higher count than the "No" category, indicating that a majority of customers opt for these services.
